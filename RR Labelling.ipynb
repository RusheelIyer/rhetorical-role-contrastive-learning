```python
import numpy as np
import torch

embedding_list = []

for doc in torch.load('datasets/embeddings.pt'):
    for sentence in doc:
        embedding_list.append(sentence.cpu().detach().numpy())
```

```python
pred_labels = []
with open(r'datasets/pred_labels.txt', 'r') as fp:
    for line in fp:
        x = line[:-1]
        pred_labels.append(x)

true_labels = []
with open(r'datasets/true_labels.txt', 'r') as fp:
    for line in fp:
        x = line[:-1]
        true_labels.append(x)

print(pred_labels)
print(true_labels)
```

```python
from sklearn.manifold import TSNE

X_embedded = TSNE(n_components=2).fit_transform(embedding_list)
X_embedded_3D = TSNE(n_components=3).fit_transform(embedding_list)
```

```python
import pandas as pd

df_labels = pd.DataFrame({'label':pred_labels})
df_labels

df_embeddings = pd.DataFrame(X_embedded)
df_embeddings = df_embeddings.rename(columns={0:'x',1:'y'})
df_embeddings = df_embeddings.assign(label=df_labels['label'])

df_embeddings
```

```python
df_embeddings_3D = pd.DataFrame(X_embedded_3D)
df_embeddings_3D = df_embeddings_3D.rename(columns={0:'x',1:'y', 2:'z'})
df_embeddings_3D = df_embeddings_3D.assign(label=df_labels['label'])

df_embeddings_3D
```

```python
import plotly.express as px

fig = px.scatter(
    df_embeddings, x='x', y='y',
    color='label', labels={'color': 'label'},
    #hover_data=['text'],
    title = 'Legal Embedding Visualization')
fig.show()
```

```python
fig = px.scatter_3d(df_embeddings_3D, x='x', y='y', z='z',
              color='label', title = 'Legal Embedding Visualization 3D',
              size_max=10)

fig.update_layout(margin=dict(l=0, r=0, b=0, t=0))
fig.update_traces(marker_size = 3)
```
